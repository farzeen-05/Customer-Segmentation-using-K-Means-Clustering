```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
annual_income=np.random.randint(20,120,42)
score=np.random.randint(1,120,42)
```

```python
x=np.column_stack([annual_income,score])
```

```python
from sklearn.preprocessing import StandardScaler
sr=StandardScaler()
y=sr.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
inertia=[]
k_means=range(1,11)
for k in k_means:
  km=KMeans(n_clusters=k,random_state=42,n_init=10)
  km.fit(y)
  inertia.append(km.inertia_)
```

```python
kme=KMeans(n_clusters=4, n_init=10, random_state=42)
```

```python
kme.fit(y)
```
